## 1. Logistic Regression

Fit a Logistic Regression using only `age` and `chol` (cholesterol) as predictors.

For a 55 year old, how high would their cholesterol need to be for the doctors to predict heart disease is present?

How high for the doctors to estimate a 90% chance that heart disease is present?

```python
X = dis[['age', 'chol']]
y = dis[['diagnosis']]

pipeline = Pipeline(
    [
     ("LR", LogisticRegression())]
)

fitted1 = pipeline.fit(X,y)

print(X.columns)

print(fitted1['LR'].coef_)

print(fitted1['LR'].intercept_)

# zi = fitted.coef_[0]*55 + fitted.coef_[1]*chol + fitted.intercept_


pi = 0.9

zi = np.log(pi/(1-pi))

chol = (zi - fitted1['LR'].coef_[0][0]*55 - fitted1['LR'].intercept_) / fitted1['LR'].coef_[0][1]

# (2.2 - 0.047*55 + 3.24) / 0.0018
print(chol)

print(zi)
```

## 2. Linear Discriminant Analysis

Fit an LDA model using only `age` and `chol` (cholesterol)  as predictors.

For a 55 year old, how high would their cholesterol need to be for the doctors to predict heart disease is present?

```python
X = dis[['age', 'chol']]
y = dis[['diagnosis']]

pipeline = Pipeline(
    [
     ("LR", LinearDiscriminantAnalysis())]
)

fitted2 = pipeline.fit(X,y)

print(X.columns)

print(fitted2['LR'].coef_)

print(fitted2['LR'].intercept_)

pi = 0.9

zi = np.log(pi/(1-pi))

chol = (zi - fitted2['LR'].coef_[0][0]*55 - fitted2['LR'].intercept_) / fitted2['LR'].coef_[0][1]

print(chol)
```

## 3. Support Vector Classifier

Fit an SVC model using only `age` and `chol` as predictors.  Don't forget to tune the regularization parameter.

For a 55 year old, how high would their cholesterol need to be for the doctors to predict heart disease is present?

```python
X = dis[['age', 'chol']]
y = dis[['diagnosis']]

pipeline = Pipeline(
    [
     ("LR", SVC(kernel = "linear"))]
)

fitted3 = pipeline.fit(X,y)

print(X.columns)

print(fitted3['LR'].coef_)

print(fitted3['LR'].intercept_)

pi = 0.9

zi = np.log(pi/(1-pi))

chol = (zi - fitted3['LR'].coef_[0][0]*55 - fitted3['LR'].intercept_) / fitted3['LR'].coef_[0][1]

print(chol)
```

## 4. Comparing Decision Boundaries

Make a scatterplot of `age` and `chol`, coloring the points by their true disease outcome.  Add a line to the plot representing the **linear separator** (aka **decision boundary**) for each of the three models above.

```python
df1 = pd.DataFrame({'y':range(30,80), 'x1':list(map(lambda x: (0 - fitted1['LR'].coef_[0][0]*x - fitted1['LR'].intercept_) / fitted1['LR'].coef_[0][1], range(30,80)))})
df1['x1'] = df1['x1'].apply(lambda x: x[0])

df2 = pd.DataFrame({'y':range(30,80), 'x2':list(map(lambda x: (0 - fitted2['LR'].coef_[0][0]*x - fitted2['LR'].intercept_) / fitted2['LR'].coef_[0][1], range(30,80)))})
df2['x2'] = df2['x2'].apply(lambda x: x[0])

df3 = pd.DataFrame({'y':range(30,80), 'x3':list(map(lambda x: (0 - fitted3['LR'].coef_[0][0]*x - fitted3['LR'].intercept_) / fitted3['LR'].coef_[0][1], range(30,80)))})
df3['x3'] = df3['x3'].apply(lambda x: x[0])
```

```python
from plotnine import (ggplot,
                     aes,
                     geom_point,
                    geom_line
                     )

(ggplot(dis, aes(x='age', y='chol'))
    + geom_point(aes(color='diagnosis'))
    + geom_line(aes(x='y', y = 'x1',color = '"Logistic Regression"'),size = 2, data = df1)
    + geom_line(aes(x='y', y = 'x2', color = '"LDA"'), data = df2)
    + geom_line(aes(x='y', y = 'x3', color = '"SVC"'), data = df3)
)

```
